## One-hot encoding using Sklearn

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder(drop='first')
```

```python
ohe.fit_transform(df[[' Not-in-family',' Bachelors']])
```

```python
df_new=ohe.fit_transform(df[[' Not-in-family',' Bachelors']]).toarray()
```

```python
df.drop(columns=[' Not-in-family',' Bachelors'])
```

```python
df.drop(columns=[' Not-in-family',' Bachelors']).values
```

```python
np.hstack((df.drop(columns=[' Not-in-family',' Bachelors']).values,df_new))
#it is having flaot values , so changing dataype 0.0
```

```python
ohe_sparse=OneHotEncoder(drop='first',sparse=False,dtype=np.int32)
```

```python
#it is having flaot values , so changing dataype
ohe_sparse.fit_transform(df[[' Not-in-family',' Bachelors']])
```

```python
df_sparse = ohe_sparse.fit_transform(df[[' Not-in-family',' Bachelors']])
```

```python
new_data = np.hstack((df.drop(columns=[' Not-in-family',' Bachelors']).values,df_sparse))
```

```python
pd.DataFrame(new_data)
```
